```python
import pandas as pd

# Load file
df = pd.read_csv("campaign_performance.csv")

# Quick look
print(df.head())
print(df.info())

# Standardize platform column
df['platform'] = df['platform'].replace({
    'insta': 'Instagram',
    'Instagram Ads': 'Instagram',
    'IG': 'Instagram',
    'fb': 'Facebook',
    'FB Ads': 'Facebook'
})

df = df.drop_duplicates()

df.to_csv("campaigns_cleaned.csv", index=False)
```

🔹 Platform Column Standardization

We noticed that the platform column contained inconsistent values such as "insta", "Instagram Ads", and "IG".
If we don’t standardize them, the same platform’s performance will get split into multiple categories, which will lead to wrong analysis.


👉 So, we standardized all variations into "Instagram".

🔹 Duplicate Removal

The dataset contained duplicate rows (same ad performance records repeated).
Duplicates can inflate ad spend or signup counts, which makes metrics incorrect.
👉 Therefore, we removed duplicates to ensure each record is unique and reliable.

🔹 Final Note

After cleaning, the dataset is consistent, free of duplicates, and ready for further analysis.
We exported this clean dataset as campaigns_cleaned.csv.
